```python
path='../input/covid19-radiography-database/COVID-19 Radiography Database'
def function():
    filenames = []
    labels = []
    covid_files_dir = path + '/COVID-19'
    pneumonia_files_dir = path + '/Viral Pneumonia'
    normal_files_dir = path + '/NORMAL'
    
    for filename in os.listdir(covid_files_dir):
        filenames.append(path + '/COVID-19/' + filename)
        labels.append(0)

    for filename in os.listdir(normal_files_dir):
        filenames.append(path + '/NORMAL/' + filename)
        labels.append(1)

    for filename in os.listdir(pneumonia_files_dir):
        filenames.append(path + '/Viral Pneumonia/' + filename)
        labels.append(2)
        
        
    df=pd.DataFrame({'Images':filenames,'Labels':labels})
    return(df)
```

```python
def get_test_and_valid_generator( test, train, x_col, y_cols, sample_size=100, batch_size=32, seed=1, target_w = 340, target_h = 340):
    raw_train_generator = ImageDataGenerator().flow_from_dataframe(
        dataframe=train,  
        x_col="Images", 
        y_col="Labels", 
        class_mode="raw", 
        batch_size=sample_size, 
        shuffle=True, 
        target_size=(target_w, target_h))
    
    # get data sample
    batch = raw_train_generator.next()
    data_sample = batch[0]

    # use sample to fit mean and std for test set generator
    image_generator = ImageDataGenerator(
        featurewise_center=True,
        featurewise_std_normalization= True)
    
    # fit generator to sample from training data
    image_generator.fit(data_sample)
    
    train_generator = image_generator.flow_from_dataframe(
            dataframe=train,
            x_col=x_col,
            y_col=y_cols,
            label_mode="float",
            class_mode="raw",
            color_mode="rgb",
            batch_size=32,
            shuffle=True,
            seed=1,
            target_size=(target_w,target_h))
    
    test_generator = image_generator.flow_from_dataframe(
            dataframe=test,
            x_col=x_col,
            y_col=y_cols,
            class_mode="raw",
            target_size=(target_w,target_h))
    return train_generator,test_generator
```

```python
from keras.preprocessing.image import ImageDataGenerator
#train_generator= get_train_generator(train,"Images",'Labels')
train_generator,test_generator= get_test_and_valid_generator(test,train,"Images",'Labels')
```

```python
import kerastuner as kt
from tensorflow import keras
import tensorflow as tf
from kerastuner.tuners import RandomSearch
from kerastuner.engine.hyperparameters import HyperParameter as hp
from keras.layers import Dense,Dropout,Activation,Add,MaxPooling2D,Conv2D,Flatten
from keras.models import Sequential 
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
from keras.applications import VGG19
from keras import layers
from keras.preprocessing import image
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
model = tf.keras.models.Sequential([
  layers.BatchNormalization(),
  layers.Conv2D(32, 3, activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, activation='relu'),
  layers.MaxPooling2D(),
  layers.Dropout(0.3),
  layers.Conv2D(128, 3, activation='relu'),
  layers.MaxPooling2D(),
  layers.Dropout(0.2),
  layers.Flatten(),
  layers.Dense(256, activation='relu'),
  layers.Dropout(0.15),
  layers.Dense(3, activation= 'softmax')
])
```

```python
early = tf.keras.callbacks.EarlyStopping(monitor='val_loss',patience=5)
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history=model.fit(train_generator,validation_data= test_generator,batch_size=32,epochs = 20,callbacks=[early])
```
